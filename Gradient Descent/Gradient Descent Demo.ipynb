```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X=np.arange(10)
Y=(X-5)**2+3
print(X,Y)
```

```python
# to visualize the data 
plt.style.use('_classic_test_patch')
plt.plot(X,Y)
plt.xlabel("x")
plt.ylabel("Y")
plt.show()
```

```python
print(plt.style.available)
```

```python
## Gradient descent 
# initialize the x variable
x=0
lr=0.1
plt.plot(X,Y)
for i in range(50):
    grad= 2*(x-5)
    x= x-lr*grad
    y=(x-5)**2 +3
    plt.scatter(x,y)
    print(x)

plt.show()
```
